## Test Hid for gamepads

```python
import hid
```

# find a controller and extract ids


```python
for device in hid.enumerate():
  print(f"0x{device['vendor_id']:04x}:0x{device['product_id']:04x} {device['product_string']}")
```

```python
devices = hid.enumerate()
controller_id = None
controller_vendor = None
for e in devices:
  # print(f"{e['product_string']}")
  if "DUALSHOCK" in e['product_string']:
    controller_id = e['product_id']
    controller_vendor = e['vendor_id']
    break

print(f"controller_id: {controller_id}")
print(f"controller_vendor: {controller_vendor}")
```

```python
gamepad = hid.device()
gamepad.open(controller_vendor, controller_id)
gamepad.set_nonblocking(True)

while True:
  if d:= gamepad.read(64):
    print(d)
```
